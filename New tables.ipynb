```python
import pandas as pd
import scipy.stats
import numpy as np
```

```python
paf = 'C:/Maxim/!GCCM/'
seazone = 4
year = 2018
```

```python
if seazone == 0:
    season = 'yy'
    ss = 'yJ-annual'
else:
    season = 's' + str(seazone)
    if season == 's1':
        ss = 's1-Winter'
        ma = 'm12'
        mb = 'm1'
        mc = 'm2'
    if season == 's2':
        ss = 's2-Spring' 
        ma = 'm3'
        mb = 'm4'
        mc = 'm5'
    if season == 's3':
        ss = 's3-Summer'
        ma = 'm6'
        mb = 'm7'
        mc = 'm8'
    if season == 's4':
        ss = 's4-Autumn' 
        ma = 'm9'
        mb = 'm10'
        mc = 'm11'
        
ye = year % 100

path = paf + 'Bulletin' + str(ye) + '/' + ss
```

```python
def Table1_1():
    igce = pd.read_table(path + '/3-Global/res_dat/' + season + '_GNS_igce.dat',\
                     sep = '\s+')
    cru = pd.read_table(path + '/3-Global/res_dat/' + season + '_GNS_cru.dat',\
                     sep = '\s+')
    
    igce.columns = ['y', 'gl_s', 'nh_s', 'sh_s', 'gl_m-1', 'nh_m-1', 'sh_m-1', 'gl_m', 'nh_m', 'sh_m', 'gl_m+1', 'nh_m+1', 'sh_m+1']
    cru.columns = ['y', 'gl_s', 'nh_s', 'sh_s', 'gl_m-1', 'nh_m-1', 'sh_m-1', 'gl_m', 'nh_m', 'sh_m', 'gl_m+1', 'nh_m+1', 'sh_m+1']
    
    cru = cru[cru['y'] > 1900]
    cru = cru.apply(pd.to_numeric)
    cru = cru.reset_index(drop=True)
    
    index = pd.MultiIndex.from_product([['1976' + '-' + str(year) + ' ' + season, '1919' + '-' + str(year) + ' ' + season], ['Gl', 'Nh', 'Sh']])
    columns = ['Corr(igce, cru)', 'e_dif', 'S_dif', 'e(abs_dif)', 'max(abs_dif)', 'b_dif*10', 'S_igce', 'S_cru', 'S(igce+cru)', 'max_dif']
    df = pd.DataFrame(index = index, columns = columns)
    
    df['Corr(igce, cru)'] = scipy.stats.pearsonr(igce['gl_s'][igce['y'] >= 1976], cru['gl_s'][cru['y'] >= 1976])[0],\
    scipy.stats.pearsonr(igce['nh_s'][igce['y'] >= 1976], cru['nh_s'][cru['y'] >= 1976])[0],\
    scipy.stats.pearsonr(igce['sh_s'][igce['y'] >= 1976], cru['sh_s'][cru['y'] >= 1976])[0],\
    scipy.stats.pearsonr(igce['gl_s'], cru['gl_s'])[0],\
    scipy.stats.pearsonr(igce['nh_s'], cru['nh_s'])[0],\
    scipy.stats.pearsonr(igce['sh_s'], cru['sh_s'])[0]
    
    dif76_Gl = igce['gl_s'][igce['y'] >= 1976] - cru['gl_s'][cru['y'] >= 1976]
    dif76_Nh = igce['nh_s'][igce['y'] >= 1976] - cru['nh_s'][cru['y'] >= 1976]
    dif76_Sh = igce['sh_s'][igce['y'] >= 1976] - cru['sh_s'][cru['y'] >= 1976]
    dif19_Gl = igce['gl_s'] - cru['gl_s']
    dif19_Nh = igce['nh_s'] - cru['nh_s']
    dif19_Sh = igce['sh_s'] - cru['sh_s']
    
    df['e_dif'] = dif76_Gl.mean(), dif76_Nh.mean(), dif76_Sh.mean(), dif19_Gl.mean(), dif19_Nh.mean(), dif19_Sh.mean()
    df['S_dif'] = dif76_Gl.std(), dif76_Nh.std(), dif76_Sh.std(), dif19_Gl.std(), dif19_Nh.std(), dif19_Sh.std()
    df['e(abs_dif)'] = dif76_Gl.abs().mean(), dif76_Nh.abs().mean(), dif76_Sh.abs().mean(),\
    dif19_Gl.abs().mean(), dif19_Nh.abs().mean(), dif19_Sh.abs().mean()
    df['max(abs_dif)'] = dif76_Gl.abs().max(), dif76_Nh.abs().max(), dif76_Sh.abs().max(),\
    dif19_Gl.abs().max(), dif19_Nh.abs().max(), dif19_Sh.abs().max()
    #df['b_dif*10'] = np.polyfit(x, y)
    df['S_igce'] = (igce['gl_s'][igce['y'] >= 1976]).std(), (igce['nh_s'][igce['y'] >= 1976]).std(), (igce['sh_s'][igce['y'] >= 1976]).std(),\
    igce['gl_s'].std(), igce['nh_s'].std(), igce['sh_s'].std()
    df['S_cru'] = (cru['gl_s'][cru['y'] >= 1976]).std(), (cru['nh_s'][cru['y'] >= 1976]).std(), (cru['sh_s'][cru['y'] >= 1976]).std(),\
    cru['gl_s'].std(), cru['nh_s'].std(), cru['sh_s'].std()
    df['S(igce+cru)'] = np.sqrt((df['S_igce'] ** 2 + df['S_cru'] ** 2) / 2)
    df['max_dif'] = dif76_Gl.max(), dif76_Nh.max(), dif76_Sh.max(), dif19_Gl.max(), dif19_Nh.max(), dif19_Sh.max()
    df = df.T
    return df
```

```python

def color_2_1(val):
    color = 'red' if val == year else 'black'
    return 'color: %s' % color
```

```python
def Table2_1():
    yr = 'y_' + season + '.'
    
    df = pd.read_table(path + '/3-Global/res_dat/GL12_ranked.dat',\
                   sep = '\s+')
    df = df[['Regions', 'reg', 'NN', season, yr]]

    df = df[df['NN'] <= 7]

    df[['Nh', 'Nh_y']] = df[[season, yr]][df['reg'] % 3 == 2]
    df[['Sh', 'Sh_y']] = df[[season, yr]][df['reg'] % 3 == 0]
    df[[season, yr]] = df[[season, yr]][df['reg'] % 3 == 1]

    df = df.rename(columns={season: 'Gl', yr: 'Gl_y'})

    Gl = df[['NN', 'Gl' ,'Gl_y']].dropna().reset_index().drop(['index'], axis=1)
    Nh = df[['NN', 'Nh', 'Nh_y']].dropna().reset_index().drop(['index'], axis=1)
    Sh = df[['NN', 'Sh', 'Sh_y']].dropna().reset_index().drop(['index'], axis=1)

    Table = Gl
    Table['Nh'] = Nh['Nh']
    Table['Nh_y'] = Nh['Nh_y']
    Table['Sh'] = Sh['Sh']
    Table['Sh_y'] = Sh['Sh_y']

    Table = Table.style.applymap(color_2_1)
    
    return Table
```
