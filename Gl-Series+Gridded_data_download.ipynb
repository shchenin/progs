```python
MMYY = 2019.01
fr = 'C:/Users/igce/progs/'
to = 'C:/Maxim/z-data_wrk/'
```

```python
import requests, shutil, os

MMYY = str(MMYY)

os.mkdir('C:/Maxim/z-data_wrk/Data_wrk_Gl-series/' + MMYY)
os.mkdir('C:/Maxim/z-data_wrk/Data_wrk_Gl-series/' + MMYY + '/data')
os.mkdir('C:/Maxim/z-data_wrk/Data_wrk_Gl-series/' + MMYY + '/data/monthly')
os.mkdir('C:/Maxim/z-data_wrk/Data_wrk_Gl-series/' + MMYY + '/data/annual')
os.mkdir('C:/Maxim/z-data_wrk/Data_wrk_Gridded/!HadCRU/' + MMYY)
os.mkdir('C:/Maxim/z-data_wrk/Data_wrk_Gridded/!HadCRU/' + MMYY + '/CRU')
os.mkdir('C:/Maxim/z-data_wrk/Data_wrk_Gridded/!HadCRU/' + MMYY + '/zipped')

GS_d = ['https://crudata.uea.ac.uk/cru/data/temperature/HadCRUT4-gl.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/HadCRUT4-nh.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/HadCRUT4-sh.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/CRUTEM4-gl.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/CRUTEM4-nh.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/CRUTEM4-sh.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/HadSST3-gl.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/HadSST3-nh.dat',\
        'https://crudata.uea.ac.uk/cru/data/temperature/HadSST3-sh.dat']

GS_d_a = ['https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.annual_ns_avg.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.annual_nh.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.annual_sh.txt',\
          'https://www.metoffice.gov.uk/hadobs/crutem4/data/diagnostics/global/nh+sh/CRUTEM.4.6.0.0.global_n+s',\
          'https://www.metoffice.gov.uk/hadobs/crutem4/data/diagnostics/hemispheric/northern/CRUTEM.4.6.0.0.nh',\
          'https://www.metoffice.gov.uk/hadobs/crutem4/data/diagnostics/hemispheric/southern/CRUTEM.4.6.0.0.sh',\
          'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/diagnostics/HadSST.3.1.1.0_annual_globe_ts.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/diagnostics/HadSST.3.1.1.0_annual_nh_ts.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/diagnostics/HadSST.3.1.1.0_annual_sh_ts.txt']

GS_d_m = ['https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.monthly_ns_avg.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.monthly_nh.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/time_series/HadCRUT.4.6.0.0.monthly_sh.txt',\
          'https://www.metoffice.gov.uk/hadobs/crutem4/data/diagnostics/global/nh+sh/CRUTEM.4.6.0.0.global_n+s_monthly',\
          'https://www.metoffice.gov.uk/hadobs/crutem4/data/diagnostics/hemispheric/northern/CRUTEM.4.6.0.0.nh_monthly',\
          'https://www.metoffice.gov.uk/hadobs/crutem4/data/diagnostics/hemispheric/southern/CRUTEM.4.6.0.0.sh_monthly',\
          'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/diagnostics/HadSST.3.1.1.0_monthly_globe_ts.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/diagnostics/HadSST.3.1.1.0_monthly_nh_ts.txt',\
          'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/diagnostics/HadSST.3.1.1.0_monthly_sh_ts.txt']


linksGridded = ['https://crudata.uea.ac.uk/cru/data/temperature/HadCRUT.4.6.0.0.median.nc',\
                'https://crudata.uea.ac.uk/cru/data/temperature/CRUTEM.4.6.0.0.anomalies.nc',\
                'https://crudata.uea.ac.uk/cru/data/temperature/HadSST.3.1.1.0.median.nc',\
                'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/gridded_fields/HadCRUT.4.6.0.0.median_ascii.zip',\
                'https://www.metoffice.gov.uk/hadobs/hadcrut4/data/current/gridded_fields/HadCRUT.4.6.0.0.median_netcdf.zip',\
                'https://www.metoffice.gov.uk/hadobs/crutem4/data/gridded_fields/CRUTEM.4.6.0.0.anomalies.txt.gz',\
                'https://www.metoffice.gov.uk/hadobs/crutem4/data/gridded_fields/CRUTEM.4.6.0.0.anomalies.nc.gz',\
                'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/ascii/HadSST.3.1.1.0.median.zip',\
                'https://www.metoffice.gov.uk/hadobs/hadsst3/data/HadSST.3.1.1.0/netcdf/HadSST.3.1.1.0.median_netcdf.zip']




for s in range(len(GS_d)):
    (dirname, filename) = os.path.split(GS_d[s])
    filename = str(s + 1) + filename + '.txt'
    r = requests.get(GS_d[s], verify=False, stream = True)
    if r.status_code == 200:
        with open(filename, 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
    shutil.move(fr + filename, to + 'z-data_wrk/Data_wrk_Gl-series/' + MMYY + '/data')

for s in range(len(GS_d_a)):
    (dirname, filename) = os.path.split(GS_d_a[s])
    filename = str(s + 1) + filename
    if (s < 3) or (s > 5):
        filename = filename[:-4] + '.txt'	
    r = requests.get(GS_d_a[s], verify=False, stream = True)
    if r.status_code == 200:
        with open(filename, 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
    shutil.move(fr + filename, to + 'Data_wrk_Gl-series/' + MMYY + '/data/annual')

for s in range(len(GS_d_m)):
    (dirname, filename) = os.path.split(GS_d_m[s])
    filename = str(s + 1) + filename
    if (s < 3) or (s > 5):
        filename = filename[:-4] + '.txt'
    r = requests.get(GS_d_m[s], verify=False, stream = True)
    if r.status_code == 200:
        with open(filename, 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
    shutil.move(fr + filename, to + 'Data_wrk_Gl-series/' + MMYY + '/data/monthly')
	
for s in range(len(linksGridded)):
    (dirname, filename) = os.path.split(linksGridded[s])
    r = requests.get(GS_d_m[s], verify=False, stream = True)
    if r.status_code == 200:
        with open(filename, 'wb') as f:
            r.raw.decode_content = True
            shutil.copyfileobj(r.raw, f)
    if s < 3:
        shutil.move(fr + filename, to + 'Data_wrk_Gridded/!HadCRU/' + MMYY + '/CRU')
    else:
        shutil.move(fr + filename, to + 'Data_wrk_Gridded/!HadCRU/' + MMYY + '/zipped')
```
